```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

# Problem1

```python
# function that returns dy/dt
def function(y,t,k):
    dydt = -k*y
    return dydt
# Initial condition
y0 = 5
t = np.linspace(0,50)
k = 0.1
y1 = odeint(function,y0,t, args = (k,))
k = 0.2
y2 = odeint(function,y0,t, args = (k,))
k = 0.5
y3 = odeint(function,y0,t, args = (k,))

# plot results
plt.plot(t,y1, 'r-' , linewidth=2, label='k=0.1')
plt.plot(t,y2, 'b--' , linewidth=2, label='k=0.2')
plt.plot(t,y3, 'g:' , linewidth=2, label='k=0.5')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

In the above example, dy/dt = -ky. This means dy/y = -kdt, integrating on both sides
lny = -kt + c, using the initial condition such that at t=0, y=y0, the solution becomes
y = y0e^(-kt). And we define various values for k.

# Problem2
dy/dt = -y(t) + 1 with y(0) = 0

```python
def problem2(y,t):
    dydt = -y + 1.0
    return dydt
# provide the initial condition
y0 = 0
#time point
t=np.linspace(0,10)
#solve ODE
y = odeint(problem2,y0,t)
#plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show
```

# Problem3

Solve ODE

\frac{dy}{dt} = ye^{-1}

# Problem4

\begin{equation}
5 \frac{dy(t)}{dt} = -y(t) + u(t)
\end{equation}
y(0) = 1

u steps from 0 to 2 at t = 10

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def problem4(y,t):
    # u steps from 0 to 2 at t=10
    if t<10.0:
        u = 0
    else:
        u = 2
    dydt = (-y + u)/5.0
    return dydt

# initial condition
y0 = 1

# time points
t = np.linspace(0,40,1000)

# solve ODE
y = odeint(problem4,y0,t)

# plot results
plt.plot(t,y,'r-',label='Output (y(t))')
plt.plot([0,10,10,40],[0,0,2,2],'b-',label='Input (u(t))')
plt.ylabel('values')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```

# Problem5

Solve for x(t) and y(t) and show that the solutions are equivalent
\begin{equation}
\frac{dx}{dt} = 3 exp(-t)\\
\frac{dy(t)}{dt} = 3 - y(t)\\
x(0) = 0\\
y(0) = 0
\end{equation}

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dz/dt
def problem5(z,t):
    dxdt = 3.0 * np.exp(-t)
    dydt = -z[1] + 3
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [0,0]

# time points
t = np.linspace(0,5)

# solve ODE
z = odeint(problem5,z0,t)

# plot results
plt.plot(t,z[:,0],'b-',label=r'$\frac{dx}{dt}=3 \; \exp(-t)$')
plt.plot(t,z[:,1],'r--',label=r'$\frac{dy}{dt}=-y+3$')
plt.ylabel('response')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```
