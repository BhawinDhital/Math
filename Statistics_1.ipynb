### Normal distribution

```python
x = np.arange(-20,20,0.01)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import math as math
sns.set()
```

### Normal distribution function :

\begin{equation}
f(x, \mu, \sigma^2 )= \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
\end{equation}



```python
def normal(x,m,s):
    f = (1/np.sqrt(2*np.pi*s**2))*np.exp(-(x-m)**2/(2*s**2))
    return f
plt.figure(figsize = [15,10])
plt.plot(x,normal(x,1.0,1.0),\
        x,normal(x,2.0,2.0),\
        x,normal(x,3.0,3.0),\
        x,normal(x,4.0,4.0),\
        x,normal(x,5.0,5.0),\
        x,normal(x,6.0,6.0))
plt.show()
```
